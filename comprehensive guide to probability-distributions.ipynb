# Bernoulli Distribution


## Comparing Population and Sample Distributions in a Bernoulli Trial

This notebook demonstrates the process of generating a Bernoulli distribution, creating frequency tables for both a large population and a smaller sample, and visualizing the results. We explore the differences and similarities between the population and sample distributions, highlighting the importance of sample size and randomness in statistical analysis.

**By following the steps outlined in this notebook, you will learn how to:**

* Define parameters for a Bernoulli distribution
* Generate a large population and a smaller sample
* Create and interpret frequency tables
* Visualize the distributions to compare the population and sample

**This practical example provides a clear understanding of how sampling works and the impact of sample size on the accuracy of statistical representations.**

```python
# Importing necessary libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from scipy.stats import t
from scipy.stats import chi2
import matplotlib.ticker as ticker
import warnings
```

## Step 1: Define Parameters

We start by defining the parameters for our Bernoulli distribution.


```python
# Probability of success
p = 0.2
# Size of the population
population_size = 100000
# Size of the sample
sample_size = 1000
```

## Step 2: Generate Bernoulli Population

We generate a large population based on the Bernoulli distribution.


```python
# Generate Bernoulli Population
population_data = np.random.binomial(1, p, population_size)
```

## Step 3: Create a Frequency Table for the Population

We create a frequency table to see the distribution of outcomes in the population.


```python
# Create a Frequency Table for the Population
population_freq = pd.Series(population_data).value_counts().sort_index()
```

## Step 4: Generate Bernoulli Sample

We take a smaller sample from the population.


```python
sample_data = np.random.choice(population_data, sample_size, replace=False)
```

## Step 5: Create a Frequency Table for the Sample

We create a frequency table to see the distribution of outcomes in the sample.


```python
sample_freq = pd.Series(sample_data).value_counts().sort_index()
```

## Step 6: Display Frequency Tables

We display the frequency tables for both the population and the sample.


```python
print("Population Frequency Table:")
print(population_freq)

print("\nSample Frequency Table:")
print(sample_freq)
```

## Step 7: Visualization

We visualize the distributions of the population and sample side by side.


```python
fig, axs = plt.subplots(1, 2, figsize=(12, 5))

# Population plot
axs[0].bar(population_freq.index, population_freq.values, color=['#FF6347', '#4682B4'])
axs[0].set_title('Population Distribution (N = 100000)')
axs[0].set_xlabel('Outcome')
axs[0].set_ylabel('Frequency')
axs[0].set_xticks([0, 1])
axs[0].set_xticklabels(["0 (Failure)", "1 (Success)"])

# Sample plot
axs[1].bar(sample_freq.index, sample_freq.values, color=['#FF6347', '#4682B4'])
axs[1].set_title('Sample Distribution (n = 1000)')
axs[1].set_xlabel('Outcome')
axs[1].set_ylabel('Frequency')
axs[1].set_xticks([0, 1])
axs[1].set_xticklabels(["0 (Failure)", "1 (Success)"])

plt.tight_layout()
plt.show()
```

# Binomial Distribution

**Comparing Population and Sample Distributions in a Binomial Trial**

This notebook demonstrates the process of generating a Binomial distribution, creating frequency tables for both a large population and a smaller sample, and visualizing the results using Seaborn. We explore the differences and similarities between the population and sample distributions, highlighting the importance of sample size and randomness in statistical analysis.

## By following the steps outlined in this notebook, you will learn how to:
*  Define parameters for a Binomial distribution
*  Generate a large population and a smaller sample
*  Create and interpret frequency tables
*  Visualize the distributions to compare the population and sample

**This practical example provides a clear understanding of how sampling works and the impact of sample size on the accuracy of statistical representations.**

## Step 1: Define Parameters

We start by defining the parameters for our Binomial distribution.


```python
# Define Parameters
# Number of trials
n_trials = 10
# Probability of success
p = 0.5
# Size of the population
population_size = 100000
# Size of the sample
sample_size = 1000
```

## Step 2: Generate Binomial Population

We generate a large population based on the Binomial distribution.


```python
# Generate Binomial Population
population_data = np.random.binomial(n_trials, p, population_size)
```

## Step 3: Create a Frequency Table for the Population

We create a frequency table to see the distribution of outcomes in the population.


```python
# Create a Frequency Table for the Population
population_freq = pd.Series(population_data).value_counts().sort_index().reset_index()
population_freq.columns = ['Number of Successes', 'Frequency']
```

## Step 4: Generate Binomial Sample

We take a smaller sample from the population.


```python
# Generate Binomial Sample
sample_data = np.random.choice(population_data, sample_size, replace=False)
```

## Step 5: Create a Frequency Table for the Sample

We create a frequency table to see the distribution of outcomes in the sample.


```python
# Create a Frequency Table for the Sample
sample_freq = pd.Series(sample_data).value_counts().sort_index().reset_index()
sample_freq.columns = ['Number of Successes', 'Frequency']
```

## Step 6: Display Frequency Tables

We display the frequency tables for both the population and the sample.


```python
# Display Frequency Tables
print("Population Frequency Table:")
print(population_freq)

print("\nSample Frequency Table:")
print(sample_freq)
```

## Step 7: Visualization with Seaborn

We visualize the distributions of the population and sample side by side using Seaborn.


```python
# Visualization with Seaborn
sns.set(style="whitegrid")
fig, axs = plt.subplots(1, 2, figsize=(14, 6))

# Population plot
sns.barplot(x='Number of Successes', y='Frequency', data=population_freq, ax=axs[0], palette='Blues_d')
axs[0].set_title('Population Distribution (N = 100000)')
axs[0].set_xlabel('Number of Successes')
axs[0].set_ylabel('Frequency')

# Sample plot
sns.barplot(x='Number of Successes', y='Frequency', data=sample_freq, ax=axs[1], palette='Blues_d')
axs[1].set_title('Sample Distribution (n = 1000)')
axs[1].set_xlabel('Number of Successes')
axs[1].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

# Normal Distribution and Z-Scores

## Comparing Population and Sample Distributions in a Normal Distribution

This notebook demonstrates the process of generating a normal distribution, calculating z-scores, and visualizing the results. We explore the differences and similarities between the population and sample distributions, highlighting the importance of standardization in statistical analysis.

By following the steps outlined in this notebook, you will learn how to:

- Define parameters for a normal distribution
- Generate a large population and a smaller sample
- Calculate z-scores for the values
- Visualize the distributions to compare the population and sample


## Step 1: Generate Data
We'll generate a sample dataset that follows a normal distribution.

```python
# Set the mean () and standard deviation ()
mean = 50
std_dev = 10

# Generate 1000 random values that follow a normal distribution
data = np.random.normal(mean, std_dev, 1000)

# Convert to a pandas DataFrame
df = pd.DataFrame(data, columns=['Value'])
```

## Step 2: Plot the Data
Let's plot the histogram of the data to visualize the normal distribution with enhanced styling.


```python
# Plot the data
plt.figure(figsize=(12, 8))
plt.hist(df['Value'], bins=30, edgecolor='black', alpha=0.7, color='skyblue', density=True)
plt.title('Histogram of Normally Distributed Data', fontsize=16, fontweight='bold')
plt.xlabel('Value', fontsize=14)
plt.ylabel('Density', fontsize=14)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.axvline(mean, color='red', linestyle='dashed', linewidth=2, label=f'Mean: {mean:.2f}')
plt.legend(fontsize=12)

# Adding KDE (Kernel Density Estimation) for smoothness
sns.kdeplot(df['Value'], color='blue', linewidth=2)
plt.annotate(f'Mean = {mean:.2f}', xy=(mean, 0.02), xytext=(mean+10, 0.025),
             arrowprops=dict(facecolor='black', shrink=0.05))

plt.show()
```

## Step 3: Calculate Z-Scores
Next, we'll calculate the z-scores for each value in the dataset.


```python
# Calculate the mean and standard deviation
mean = df['Value'].mean()
std_dev = df['Value'].std()

# Calculate the z-scores
df['Z-Score'] = (df['Value'] - mean) / std_dev
```

## Step 4: Plot the Z-Scores
Finally, we'll plot the histogram of the z-scores to visualize the standard normal distribution.


```python
# Plot the histogram of z-scores
plt.figure(figsize=(12, 8))
plt.hist(df['Z-Score'], bins=30, edgecolor='black', alpha=0.7, color='lightgreen', density=True)
plt.title('Histogram of Z-Scores', fontsize=16, fontweight='bold')
plt.xlabel('Z-Score', fontsize=14)
plt.ylabel('Density', fontsize=14)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.axvline(0, color='red', linestyle='dashed', linewidth=2, label='Mean: 0')
plt.legend(fontsize=12)

# Adding KDE (Kernel Density Estimation) for smoothness
sns.kdeplot(df['Z-Score'], color='blue', linewidth=2)
plt.annotate('Mean = 0', xy=(0, 0.25), xytext=(1, 0.3),
             arrowprops=dict(facecolor='black', shrink=0.05))

plt.show()
```

# Student's t-Distribution 

## Exploring the Student's t-Distribution

This notebook demonstrates the process of generating data from a Student's t-distribution and visualizing the results. We will explore the properties of the t-distribution, which is particularly useful for small sample sizes in statistical analysis.

By following the steps outlined in this notebook, you will learn how to:

- Generate sample data using the Student's t-distribution
- Define degrees of freedom for the t-distribution
- Visualize the t-distribution and understand its probability density function


## Step 1: Generate Sample Data using t-Distribution

We will generate sample data using the `np.random.standard_t` function. This function generates random samples from the standard Student's t-distribution with specified degrees of freedom.

```python
# Degrees of freedom
df_t = 10

# Generating a sample data of 15 values from a t-distribution with specified degrees of freedom
np.random.seed(42)
sample_data = np.random.standard_t(df_t, size=15)

# Creating a DataFrame for better visualization
df = pd.DataFrame(sample_data, columns=['Sample Data'])
df
```

## Step 2: Plotting the t-Distribution

We will plot the t-distribution for the specified degrees of freedom.


```python
# Define x range for the plot
x = np.linspace(-4, 4, 1000)

# Plot the t-distribution
plt.plot(x, t.pdf(x, df=df_t), label=f't-distribution df={df_t}')
plt.title("Student's t-Distribution")
plt.xlabel('t-value')
plt.ylabel('Probability Density')
plt.legend()
plt.show()
```

## Summary

In this notebook, we have:
1. Generated sample data using `np.random.standard_t`.
2. Plotted the t-distribution.

# Chi-Squared Distribution: An Introduction

## Exploring the Chi-Squared Distribution

This notebook demonstrates the process of generating data from a Chi-Squared distribution and visualizing the results. We will explore the properties of the Chi-Squared distribution, which is often used in hypothesis testing and confidence interval estimation.

By following the steps outlined in this notebook, you will learn how to:

- Generate sample data using the Chi-Squared distribution
- Define degrees of freedom for the Chi-Squared distribution
- Visualize the Chi-Squared distribution and understand its probability density function


## Step 1: Generate Sample Data using Chi-Squared Distribution

We will generate sample data using the `np.random.chisquare` function. This function generates random samples from the Chi-Squared distribution with specified degrees of freedom.

```python
warnings.filterwarnings('ignore', category=FutureWarning)

# Degrees of Freedom
df = 5  # For a fair six-sided die

# Generate Chi-squared distribution data
data_chi_squared = np.random.chisquare(df, 1000)

# Calculate the mean of the distribution
actual_mean = df
```

## Step 2: Plotting the Chi-Squared Distribution

We will plot the Chi-Squared distribution for the specified degrees of freedom.


```python
# Set the dark style
sns.set_style("darkgrid")

# Set up the figure and axes for a cleaner layout
plt.figure(figsize=(12, 7))
ax = sns.histplot(data_chi_squared, kde=True, color='deepskyblue', binwidth=.5, edgecolor='navy', 
                  linewidth=1.5, stat="probability", label=None)

# Add shading under the curve
kde_x, kde_y = ax.lines[0].get_data()
ax.fill_between(kde_x, kde_y, color="skyblue", alpha=0.5, label='KDE')

# Add a vertical line for the actual mean of the Chi-squared distribution
plt.axvline(x=actual_mean, color='darkred', linestyle='--', linewidth=2, label=f'Mean: {actual_mean:.2f}')

# Improve the title and axis labels
plt.title('Chi-squared Distribution - Goodness-of-Fit Testing', fontsize=16)
plt.xlabel('Chi-squared Value', fontsize=14)
plt.ylabel('Probability', fontsize=14)

# Customize the ticks for better readability
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# Format y-axis to show probability with two decimal places
ax.yaxis.set_major_formatter(ticker.FormatStrFormatter('%.2f'))

# Optional: Add grid lines for better visual comparison
plt.grid(True, which='both', linestyle='--', linewidth=0.5)

# Enhance with a legend
plt.legend(loc='upper right')

# Show the plot
plt.show()
```

# Poisson Distribution

## Comparing Population and Sample Distributions in a Poisson Process

This notebook demonstrates the process of generating a Poisson distribution, creating frequency tables for both a large population and a smaller sample, and visualizing the results. By following the steps outlined in this notebook, you will learn how to:

2. Generate a large population and a smaller sample.
3. Create and interpret frequency tables.
4. Visualize the distributions to compare the population and sample.



### Step 1: Define Parameters for a Poisson Distribution
First, we'll define the rate parameter () for our Poisson distribution. This parameter represents the average number of events per interval.

```python
# Set the lambda () parameter
lambda_param = 5

# Generate 1000 random values that follow a Poisson distribution
data = np.random.poisson(lambda_param, 1000)

# Convert to a pandas DataFrame
df = pd.DataFrame(data, columns=['Value'])
```

## Step 2: Plot the Data
Let's plot the histogram of the data to visualize the Poisson distribution with enhanced styling.


```python
# Plot the histogram of Poisson-distributed data
plt.figure(figsize=(10, 6))
plt.hist(df['Value'], bins=range(min(df['Value']), max(df['Value']) + 2), edgecolor='black', alpha=0.7, color='skyblue', density=True)
plt.title('Histogram of Poisson Distributed Data', fontsize=15)
plt.xlabel('Value', fontsize=13)
plt.ylabel('Density', fontsize=13)
plt.grid(True)
plt.show()
```

# F-Distribution

The F-distribution is a continuous probability distribution that arises frequently in the analysis of variance (ANOVA) and in regression analysis. It is used to compare variances between groups and is crucial in hypothesis testing, particularly for comparing statistical models.

### Properties:
- The F-distribution is skewed to the right.
- It is defined by two sets of degrees of freedom: \( d_1 \) (numerator degrees of freedom) and \( d_2 \) (denominator degrees of freedom).
- As the degrees of freedom increase, the F-distribution approaches a normal distribution.

### Real-World Application
The F-distribution is often used in the following scenarios:
- ANOVA (Analysis of Variance):** To test if there are significant differences between the means of three or more groups.
- Regression Analysis:** To test the overall significance of a regression model.
- Comparing Variances:** To compare the variances of two independent samples.

```python
# Parameters
d1 = 5  # degrees of freedom for the numerator
d2 = 20  # degrees of freedom for the denominator

# Generate data
data = np.random.f(d1, d2, 1000)

# Create a DataFrame
df = pd.DataFrame(data, columns=['F-Distribution'])

# Plotting
plt.figure(figsize=(10, 6))
df.plot.kde()
plt.title('F-Distribution')
plt.xlabel('Value')
plt.ylabel('Density')
plt.show()
```
